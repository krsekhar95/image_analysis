```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
from PIL import Image
import requests
from io import BytesIO
```

```python
url = 'https://upload.wikimedia.org/wikipedia/en/7/7d/Lenna_%28test_image%29.png'
```

```python
r = requests.get(url)
```

```python
r.status_code
```

```python
img = Image.open(BytesIO(r.content))
```

```python
img = np.array(img)
```

```python
img.shape
```

```python
#Y = 0.2125 R + 0.7154 G + 0.0721 B
img_gray = np.average(img,axis = 2,weights=[0.2125,0.7154,0.0721])
```

```python
fig,ax = plt.subplots(1,2)
ax[0].imshow(img)
ax[0].set(title = 'color image')
ax[1].imshow(img_gray,cmap = 'gray')
ax[1].set(title ='Gray Image')
plt.show()
```
